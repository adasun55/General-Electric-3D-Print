```python
'''
This notebook loads .csv in a folder, joins them into one df and 
exports everything into one final csv

Used to process the area/gradient features into 
one csv
'''
```

```python
import pandas as pd
import glob
folder_name = '182_grad_v3'
path = r'.\182_grad_v3'# use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame
```

```python
frame = frame.drop(columns='Unnamed: 0')
```

```python
frame.to_csv('182_grad_v3.csv', index=False) 
```
